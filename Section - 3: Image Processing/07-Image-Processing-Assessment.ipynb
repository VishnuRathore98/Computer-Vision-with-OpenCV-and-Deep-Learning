# Image Processing Assessment

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

**TASK: Open and display the giaraffes.jpg image that is located in the DATA folder.**

```python
giaraffes = cv2.imread("../DATA/giraffes.jpg")
show_giaraffes = cv2.cvtColor(giaraffes, cv2.COLOR_BGR2RGB)
```

```python
display_img(show_giaraffes)
```

**TASK:Apply a binary threshold onto the image.**

```python
gray_giaraffes = cv2.cvtColor(giaraffes, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray_giaraffes,127,255, cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh, cmap='gray')
```

**TASK: Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)**

```python
giaraffes = cv2.imread("../DATA/giraffes.jpg")
show_giaraffes_rgb = cv2.cvtColor(giaraffes, cv2.COLOR_BGR2RGB)
kernel = np.ones((4,4), np.float32)*0.1
```

```python
kernel
```

```python
blurred_giaraffes = cv2.filter2D(show_giaraffes_rgb,-1,kernel)
display_img(blurred_giaraffes)
```

**TASK: Create a Horizontal Sobel Filter (sobelx from our lecture) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.**

```python
giaraffes_gray = cv2.imread("../DATA/giraffes.jpg",0)
sobelx = cv2.Sobel(giaraffes_gray,cv2.CV_32F,1,0,ksize=5)
display_img(sobelx,'gray')
```

**TASK: Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.**

```python
giaraffes = cv2.imread("../DATA/giraffes.jpg")
color = ('b', 'g', 'r')
for i, c in enumerate(color):
    hist = cv2.calcHist([giaraffes], [i], None, [256], [0,256])
    plt.plot(hist, color=c)
    plt.xlim([0,256])
```
